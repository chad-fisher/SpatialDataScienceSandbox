# About Google Earth Engine

Google Earth Engine combines a multi-petabyte catalog of satellite imagery and geospatial datasets with planetary-scale analysis capabilities. Scientists, researchers, and developers use Earth Engine to detect changes, map trends, and quantify differences on the Earth's surface. It is a powerful source of pre-processed data that can easily be filtered, manipulated, and analyzed in the cloud.

You can also use Python and some scripting to access this huge catalog and download it for use in ArcGIS Pro or another system. That is what we will focus on in this tutorial.
